```python
cast_df = pd.read_csv('/Users/parkerharalds/Downloads/title.crew.tsv', delimiter='\t')
```

```python
name_df = pd.read_csv('/Users/parkerharalds/Downloads/name.basics.tsv', delimiter='\t')
```

Will now merge df together to get main_df

```python
movies = pd.merge(movie_title,ratings_df,on='tconst')
```

```python
movie_df = pd.merge(movies,cast_df,on='tconst')
```

```python
movie_df
```

```python
#renaming directors column for merge
movie_df = movie_df.rename(columns={'directors':'nconst'})
```

```python
main_df = pd.merge(movie_df,name_df,on='nconst')
```
